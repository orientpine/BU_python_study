## 학습목표
 1. pivot, pivot_table 함수의 이해

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    '이름': ['손흥민', '케인','호날두','음바페', '메시'],
    '팀' : ['토트넘', '토트넘', '맨유', '파리', '파리'],
    '등번호': ['7', '10', '7', '10', '30'],
    '득점 수': ['23', '18', '20', '23', '8']
    })

df
```

#### pivot 
 - dataframe의 형태를 변경
 - 인덱스, 컬럼, 데이터로 사용할 컬럼을 명시

```python
df.pivot('이름', '팀')
```

#### pivot_table
 - 기능적으로 pivot과 동일
 - pivot과의 차이점
   - 중복되는 모호한 값이 있을 경우, aggregation 함수 사용하여 값을 채움

```python
pd.pivot_table(df, index='이름', columns='팀', aggfunc=np.mean)
```
