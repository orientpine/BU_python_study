### 학습목표
 - 다양한 방법으로 ndarray 생성하기

```python
import numpy as np
```

#### np.array 함수로 생성하기

```python
x = np.array([1, 2, 3, 4])
print(x)

y = np.array([[2, 3, 4], [1, 2, 5]])
print(y)

print(type(y))
```

#### np.arange 함수로 생성하기

```python
np.arange(10)
```

```python
np.arange(1, 10)
```

```python
np.arange(1, 10, 2)
```

```python
np.arange(5, 101, 5)
```

#### np.ones, np.zeros로 생성하기

```python
np.ones((4, 5))
```

```python
np.ones((2, 3, 4))
```

```python
np.zeros((5,2,3,8,8))
```

#### np.empty, np.full로 생성하기

```python
np.empty((2, 3, 4))
```

```python
np.full((3, 4, 2), 7)
```

#### np.eye로 생성하기
 - 단위 행렬 생성

```python
np.eye(5)
```

#### np.linspace로 생성하기

```python
np.linspace(1, 10, 3)
```

#### reshape 함수 활용
 - ndarray의 형태, 차원을 바꾸기 위해 사용

```python
x = np.arange(1, 16)
print (x)
x.shape
x.reshape(5, 3)
```
