## 학습목표
 1. stack, unstack 함수 이해하기

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    '이름': ['손흥민', '케인','호날두','음바페', '메시'],
    '팀' : ['토트넘', '토트넘', '맨유', '파리', '파리'],
    '등번호': ['7', '10', '7', '10', '30'],
    '득점 수': ['23', '18', '20', '23', '8']
    })
df
```

####  stack & unstack
 - stack : 컬럼 레벨에서 인덱스 레벨로 dataframe 변경
  - 즉, 데이터를 쌓아올리는 개념으로 이해하면 쉬움
 - unstack : 인덱스 레벨에서 컬럼 레벨로 dataframe 변경
  - stack의 반대 operation
 
 - 둘은 역의 관계에 있음

```python
new_df = df. set_index(['이름' , '등번호'])
new_df
```

```python
new_df.unstack(0)
```

```python
new_df.unstack(1)
```

```python
# 첫번째 레벨의 컬럼을 인덱스로 이동
new_df.unstack(0).stack(0)
```
